```python
import matplotlib.pyplot as plt

# Step labels
steps = ['Dummy', 'KNN Training', 'KNN Prediction', 'KNN Retraining', 'KNN Retrained Prediction']

# Actual values from the output
accuracy = [0.5075, 0.895, 0.963, 0.721, 0.965]
recall =   [0.5048, 0.8867, 0.9407, 0.4163, 0.9130]
precision = [0.5327, 0.9045, 0.9061, 1.0, 0.8873]

# Plot
plt.figure(figsize=(10,6))
plt.plot(steps, accuracy, marker='o', label='Accuracy')
plt.plot(steps, recall, marker='s', label='Recall')
plt.plot(steps, precision, marker='^', label='Precision')
plt.title('Model Performance Over Steps')
plt.xlabel('Steps')
plt.ylabel('Score')
plt.ylim(0, 1.05)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
